#### 수집 URL 구조
- https://pcmap.place.naver.com/place/{place_id}/home
- place_id → 원본 데이터의 place_id 컬럼

```python
import pandas as pd
df= pd.read_csv('소상공인시장진흥공단_상가정보_추가수집데이터.csv')
```

```python
# brand가 '역전할머니' 인것만 출력
result = df[df['brand'] == '역전할머니']
```

```python
# brand가 '역전할머니' 인것만 파일로 저장
result.to_csv('역전할머니.csv', index=False)
```
