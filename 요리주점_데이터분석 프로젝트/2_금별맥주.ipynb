### ***중복처리 순서***
1. **중복 place_id 처리**
    - place_id_fix 사전을 기준으로 중복 place_id 중 올바른 지점만 남기고 나머지는 NaN 처리
    - 잘못된 중복을 제거하기 전에 모든 행을 살려두는 것이 중요

2. **신규 지점 place_id 지정**

    2-1.**처음 데이터에는 중복 place_id가 있음**
    - 예: 1035417750이라는 place_id가 구로디지털단지점과 신도림점에 동시에 존재
    - 즉, 중복 때문에 원래 의도한 지점과 일치하지 않음

    2-2. **place_id를 확인/검색해서 정확한 지점 정보 확보**
    - 예를 들어 신도림점은 실제 place_id가 1189283598임 → 기존 중복 place_id와 다름
    
    2-3. **정확한 매칭을 위해 기존 place_id를 교체**
    - 기존 place_id가 잘못 들어간 행 → 새로 확인한 정확한 place_id로 변경
    - 이렇게 해야 크롤링이나 분석할 때 실제 매장과 데이터가 일치

    2-4. **데이터 일관성 확보**
    - 모든 지점이 고유 place_id를 갖게 되면
    - 중복/오류로 인한 혼동 제거
    - NaN 또는 잘못된 데이터 없이 분석 가능

3. **place_id dtype 통일**
    - NaN이 존재할 수 있으므로 **nullable integer (Int64)**로 변환

4. **폐점 및 잘못된 중복 행 제거**
    - place_id가 NaN인 행 → 중복 처리 실패 또는 존재하지 않는 place_id
    - 폐점 지점 제거 (예: 안산중앙점)

5. **최종 확인**
    - 모든 place_id가 **정수(Int64)**이고
    - 중복 없이 올바른 지점만 남았는지 확인

### ***금별맥주2 내용 합치기***
- 금별맥주2 + 금별맥주4 = 전체 데이터
- 금별맥주2에서 중복 place_id 제거(금별맥주3에 포함된 행 제외) 후 금별맥주 4와 나머지 합치기
- 이렇게 하면 중복 place_id만 작업한 금별맥주3 + 나머지 금별맥주2가 합쳐져서 전체 데이터셋이 됨
- 중복 처리 및 NaN 삭제가 반영된 완전한 금별맥주4 완성

### ***검증***
- 중복 place_id가 잘 처리됐는지 확인

```python
# 금별맥주3(place_id가 중복)에서 지점명이 NaN이거나 지점명 "중앙점"인 행 삭제
삭제대상 = 금별맥주3.loc[
    금별맥주3['지점명'].isna() | (금별맥주3['지점명'] == "중앙점")
]
```

```python
# 삭제 후 금별맥주 4에 저장
# 원본 금별맥주3은 그대로 유지되고, 삭제된 복사본이 금별맥주4가 됨
금별맥주4 = 금별맥주3.drop(삭제대상.index)

삭제대상[['place_id', '상호명', '지점명']]
```

```python
# 지점명이 NaN이거나 '중앙점'인 행이 남아있는지 확인
금별맥주4.loc[금별맥주4['지점명'].isna() | (금별맥주4['지점명'] == "중앙점")]
```

```python
# 중복 place_id가 있는지 확인
금별맥주4['place_id'].duplicated().sum()
```

```python
# 중복 place_id 확인
금별맥주4.loc[
    금별맥주4['place_id'].duplicated(keep=False)  # 중복 place_id 모두 선택
].sort_values('place_id')[  # place_id 기준으로 정렬
    ['place_id', '상호명', '지점명','지번주소', '도로명주소','위도','경도']  # 확인할 컬럼 선택
]
```

```python
# 중복 place_id 재확인
금별맥주4.loc[
    금별맥주4['place_id'].duplicated(keep=False)  # 중복 place_id 모두 선택
].sort_values('place_id')[  # place_id 기준으로 정렬
    ['place_id', '상호명', '지점명','지번주소', '도로명주소','위도','경도']  # 확인할 컬럼 선택
]

# 중복 place_id 없는거 확인 완료
```

```python
# 중복 place_id 삭제 한 데이터 (지점명 NaN, 중앙점은 금별맥주3에서 삭제함)
금별맥주4[['place_id', '상호명', '지점명', '지번주소', '도로명주소', '위도', '경도']]
```

### ***현재 목표***
- 금별맥주2 + 금별맥주4 = 정제완료 한 금별맥주5 만들기
- 중복 place_id 없어야 함
- 일부 지점명이 NaN인건 괜찮음

- 금별맥주2
    - 원본 데이터 + 일부 중복 place_id 포함
    - 중복 place_id를 추출해서 금별맥주3에서 정제 작업 시작

- 금별맥주4
    - 금별맥주3에서 중복 place_id 중 불필요한 행 삭제 + place_id 수정 완료
    - 즉, 중복 place_id 정제 완료본

**- 주의할 점**
- 금별맥주4는 금별맥주2에서 추출한 행들을 정제한 결과물
- 따라서 금별맥주2에서 금별맥주4에서 처리한 동일한 행(같은 place_id) 을 그대로 두면 중복 발생
- 그래서 금별맥주2에서 금별맥주4에 포함된 place_id 행을 제거하고 합쳐야 함 → 중복 방지

```python
# 금별맥주4에 있는 place_id를 금별맥주2에서 제거
금별맥주2_remaining = 금별맥주2[~금별맥주2['place_id'].isin(금별맥주4['place_id'])]
```

```python
# 남은 금별맥주2 + 정제된 금별맥주4 합치기
금별맥주5 = pd.concat([금별맥주2_remaining, 금별맥주4], ignore_index=True)
```

```python
# 중복 검증
print("place_id 중복 검증(0이면 중복 제거 완료):", 금별맥주5['place_id'].duplicated().sum())  # 0이면 중복 제거 완료
```

```python
행개수 = 금별맥주5.shape[0]
print("금별맥주5(정제완료) 행 개수:",행개수)

금별맥주5[['place_id', '상호명', '지점명', '지번주소', '도로명주소', '위도', '경도']]
```
