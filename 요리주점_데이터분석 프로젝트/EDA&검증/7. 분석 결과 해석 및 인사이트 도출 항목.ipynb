### **7. 분석 결과 해석 및 인사이트 도출 (데이터 기반)**

#### **1️⃣ 자사 vs 경쟁사 성과 우위 구조 확정**
**1. 성과지표별 우위 브랜드 판정**
- **방문자 리뷰 수**
    - KDE, 박스플롯 결과, 밀도 그룹별 비교에서 자사/경쟁사 중 중앙값이 높은 브랜드를 우위로 판정
    - 예: 월평균_리뷰수_log 기준, 중앙값과 상위 75% 분위값 비교
- **블로그 리뷰 수**
    - 동일 시군구/밀도 조건에서 평균값과 상위 분위값 비교
    - 특정 입지에서 경쟁사보다 자사가 월등히 많은 블로그 리뷰를 확보하면 브랜드 우위로 기록
- **월평균 리뷰 수 (log)**
    - 입지 통제 후 성과 분포를 통해 상대적 안정성과 우위 판정 가능
    
**2. 우위의 크기 확정 (효과크기 기반)**
- **단순 평균 비교가 아닌 효과크기(Cohen’s d, Glass’ delta) 적용**
    - d > 0.2: 작은 차이, d > 0.5: 중간 차이, d > 0.8: 큰 차이
- 예: 자사 방문자 리뷰 수 평균 > 경쟁사 평균, d = 0.6 → 중간 수준 우위

#### **2️⃣ 동일 입지 조건에서의 브랜드 효과**
**3. 입지 통제 후 성과 차이 유지 여부**
- 시군구 단위, 요리주점 밀도 그룹, 인구밀도 그룹 등 입지 변수를 통제한 후 성과 비교
- KDE/Boxplot 결과에서 브랜드 간 차이가 유지되면 브랜드 자체 효과 존재

**4. 브랜드 효과 vs 입지 효과 구분**
- 회귀분석 또는 ANCOVA 가능 (종속: 방문자 리뷰 수, 독립: 입지 변수 + 브랜드)
    - 브랜드 계수 유의 → 브랜드 자체 효과 존재
    - 입지 계수 유의 → 입지 효과 존재
- 효과 크기 비교: 어떤 요소가 성과 변동에 더 큰 기여를 하는지 정량화

#### **3️⃣ 입지 민감도 차이 확정**
**5. 브랜드별 입지 민감도 비교**
- 요리주점 밀도 대비 성과 변화율 확인
    - High 밀도에서 리뷰 감소 폭이 큰 브랜드 → 입지 민감도가 높은 브랜드
    - Low/Mid 밀도에서 안정적 성과 → 입지 민감도가 낮은 브랜드

**6. 입지 전략 방향성 확정**
- 민감도가 낮은 브랜드 → 주요 상권 중심 출점 전략 가능
- 민감도가 높은 브랜드 → 입지 경쟁 최소화 전략 필요, 신규 점포 선정 시 경쟁 강도 회피

#### **4️⃣ 상호작용 기반 전략 분기**
**7. 상호작용 존재 여부 판정**
- 입지 변수 × 브랜드 간 상호작용 검정 (예: 두 요인 ANOVA)
- 상호작용 유의 → 브랜드 성과가 입지 조건에 따라 달라짐

**8. 전략 분기 규칙 도출**
- High 밀도 + 경쟁사 점포 근접 → 민감 브랜드는 출점 지양
- Low/Mid 밀도 → 브랜드 효과 우위 점포 적극 확장

#### **5️⃣ 실행 가능한 결론 형태로 변환**
**9. 입지 통제 후 남는 격차를 실행 과제로 번역**
- 브랜드 자체 우위가 입지와 무관하게 존재 → 마케팅·콘텐츠 강화 우선
- 입지 민감도가 높으면 → 입지 분석 자동화, 경쟁 점포 회피 로직 필요

**10. 입지 반응 구조를 출점·타깃 규칙으로 변환**
- 입지 변수 범주화 → 출점 우선순위 매트릭스 생성
    - 예: Low/Mid 밀도 + 자사 우위 → 신규 점포 타깃
    - High 밀도 + 경쟁사 점포 인접 → 출점 최소화
- 브랜드별 민감도 차이를 고려한 맞춤형 출점 전략 수립

### **정리 요약**
**1. 리뷰 수·월평균 리뷰 수 기준으로 자사 vs 경쟁사 우위 판정 가능**

**2. 입지 변수 통제 후 브랜드 효과 존재 여부 확인 → 전략적 출점 판단 근거 확보**

**3. 입지 민감도 기반으로 브랜드별 출점·타깃 전략 차별화**

**4. 상호작용 검정을 통해 입지×브랜드 성과 패턴 도출 → 실행 가능한 출점 규칙 설계 가능**

```python
import pandas as pd
import numpy as np
from scipy import stats
import statsmodels.api as sm
import statsmodels.formula.api as smf
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib import font_manager, rc

# ==============================
# 한글 폰트 설정
# ==============================
plt.rcParams['font.family'] = 'Malgun Gothic'  # 한글폰트 적용 (Windows 기준)
plt.rcParams['axes.unicode_minus'] = False

# ==============================
# 1. 데이터 불러오기
# ==============================
path1 = r"C:\Users\user\Desktop\260219_데이터분석\분석용\merge_자사경쟁사.csv"
path2 = r"C:\Users\user\Desktop\260219_데이터분석\분석용\Step2_입지변수_결합완료.csv"

df1 = pd.read_csv(path1)
df2 = pd.read_csv(path2)

# ==============================
# 2. place_id 기준 병합
# ==============================
df1['place_id'] = df1['place_id'].astype(str)
df2['place_id'] = df2['place_id'].astype(str)

df = df2.merge(
    df1[['place_id','월평균_리뷰수_log','brand_type','요리주점_존재']],
    on='place_id',
    how='left'
)

# ==============================
# 3. 결측 처리
# ==============================
df['요리주점_밀도'] = pd.to_numeric(df['요리주점_밀도'], errors='coerce')
df['방문자_리뷰수'] = pd.to_numeric(df['방문자_리뷰수'], errors='coerce')
df['블로그_리뷰수'] = pd.to_numeric(df['블로그_리뷰수'], errors='coerce')
df['월평균_리뷰수_log'] = pd.to_numeric(df['월평균_리뷰수_log'], errors='coerce')

df = df.dropna(subset=['요리주점_밀도','방문자_리뷰수','블로그_리뷰수','월평균_리뷰수_log','brand_type'])
df['시군구명'] = df['시군구명'].str.strip()

print("최종 분석용 df shape:", df.shape)
print(df['brand_type'].value_counts())

# ==============================
# 4. 성과지표별 브랜드 우위 & 효과크기
# ==============================
performance_cols = ['방문자_리뷰수', '블로그_리뷰수', '월평균_리뷰수_log']

# 4-1. 브랜드별 평균/중앙값
brand_summary = df.groupby('brand_type')[performance_cols].agg(['mean','median'])
print("\n브랜드별 성과 요약:\n", brand_summary)

# 4-2. 효과크기 계산 (Cohen's d)
def cohen_d(x, y):
    nx = len(x)
    ny = len(y)
    dof = nx + ny - 2
    pooled_std = np.sqrt(((nx-1)*np.std(x, ddof=1)**2 + (ny-1)*np.std(y, ddof=1)**2)/dof)
    return (np.mean(x) - np.mean(y)) / pooled_std

effect_sizes = {}
brands = df['brand_type'].unique()
for col in performance_cols:
    x = df[df['brand_type'] == brands[0]][col].dropna()
    y = df[df['brand_type'] == brands[1]][col].dropna()
    effect_sizes[col] = cohen_d(x, y)

print("\n효과크기 (Cohen's d) by 성과 지표:", effect_sizes)

# ==============================
# 5. 입지 통제 후 성과 차이 (ANCOVA)
# ==============================
for col in performance_cols:
    formula = f"{col} ~ C(brand_type) + 요리주점_밀도"
    model = smf.ols(formula, data=df).fit()
    print(f"\n=== {col} ANCOVA 결과 ===")
    print(model.summary())

# ==============================
# 6. 입지 민감도
# ==============================
df['밀도그룹'] = pd.qcut(df['요리주점_밀도'], q=3, labels=['Low','Mid','High'])
density_sensitivity = df.groupby(['brand_type','밀도그룹'])[performance_cols].mean().unstack(0)
print("\n입지 민감도 (밀도 그룹별 평균 성과):\n", density_sensitivity)

# 시각화
for col in performance_cols:
    plt.figure(figsize=(8,6))
    sns.lineplot(
        data=df,
        x='밀도그룹',
        y=col,
        hue='brand_type',
        marker='o'
    )
    plt.title(f'입지 민감도 비교: {col}')
    plt.ylabel('평균 성과')
    plt.xlabel('요리주점 밀도 그룹')
    plt.show()

# ==============================
# 7. 상호작용 검정 (입지 x 브랜드)
# ==============================
for col in performance_cols:
    formula = f"{col} ~ C(brand_type) * 밀도그룹"
    model = smf.ols(formula, data=df).fit()
    anova_table = sm.stats.anova_lm(model, typ=2)
    print(f"\n=== {col} 상호작용 검정 (입지 x 브랜드) ===")
    print(anova_table)
```
